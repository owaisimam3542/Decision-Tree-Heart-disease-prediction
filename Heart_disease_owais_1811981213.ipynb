```python
# Load libraries
import pandas as pd
from sklearn.tree import DecisionTreeClassifier # Import Decision Tree Classifier
from sklearn.model_selection import train_test_split # Import train_test_split function
from sklearn import metrics #Import scikit-learn metrics module for accuracy calculation

col_names = ['age','gender','chest_pain','rest_bps','cholestrol','fasting_blood_sugar','rest_ecg','thalach','exer_angina','old_peak','slope','ca','thalassemia','target']
# load dataset
pima = pd.read_csv("C:\\Users\hp\Downloads\\HeartDisease.csv", header=1, names=col_names)
pima.head()

#split dataset in features and target variable
feature_cols = ['age','gender','chest_pain','rest_bps','cholestrol','fasting_blood_sugar','rest_ecg','thalach','exer_angina','old_peak','slope','ca','thalassemia']
X = pima[feature_cols] # Features
y = pima['target'] # Target variable

# Split dataset into training set and test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1) # 70% training and 30% test

# Create Decision Tree classifer object
clf = DecisionTreeClassifier()

# Train Decision Tree Classifer
clf = clf.fit(X_train,y_train)

#Predict the response for test dataset
y_pred = clf.predict(X_test)


# Model Accuracy, how often is the classifier correct?
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
from sklearn import tree

clf = tree.DecisionTreeClassifier()                                             

#[height, hair-length, voice-pitch]                                             

clf = clf.fit(X_train, y_train)
u=int(input("Enter age"))
b=int(input("Enter gender"))
c=int(input("Enter chest_pain"))
d=int(input("Enter rest_rbps"))
e=int(input("Enter cholestrol"))
f=int(input("Fasting blood chlosterol"))
g=int(input("rest ecg"))
h=int(input("thalach"))
o=int(input("enter exer"))
k=int(input("enter old peak"))
l=int(input("enter slope"))
p=int(input("enter ca"))
u=int(input("enter tahal"))
prediction = clf.predict([[u,b,c,d,e,f,g,h,o,k,l,p,u]])                                         
print("Prediction of target{0}".format(prediction))
from sklearn.tree import export_graphviz
from six import StringIO  
from IPython.display import Image  
import pydotplus
dot_data = StringIO()
export_graphviz(clf, out_file=dot_data,  
                filled=True, rounded=True,
                special_characters=True,feature_names = feature_cols,class_names=['0','1'])
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())  
graph.write_png('h1.png')
Image(graph.create_png())

# Create Decision Tree classifer object
clf = DecisionTreeClassifier(criterion="entropy", max_depth=3)

# Train Decision Tree Classifer
clf = clf.fit(X_train,y_train)

#Predict the response for test dataset
y_pred = clf.predict(X_test)

# Model Accuracy, how often is the classifier correct?
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))

from six import StringIO  
from IPython.display import Image  
from sklearn.tree import export_graphviz
import pydotplus
dot_data = StringIO()
export_graphviz(clf, out_file=dot_data,  
                filled=True, rounded=True,
                special_characters=True, feature_names = feature_cols,class_names=['0','1'])
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())  
graph.write_png('h1.png')
Image(graph.create_png())


      

                                         
```
